```python
#Exercício 3
#1. Crie um programa que contenha um dicionário com o nome
#do aluno e a nota dele.
#2. Imprima a lista dos alunos aprovados, iterando sobre os itens
#do dicionário.
#Obs: o aluno é aprovado quando a nota for maior ou igual a 7.0

dictAlunos = {
    'Sara': 6,
    'Helo': 9,
    'Caio': 8,
    'Beatriz': 5
}

aprovados = []
for aluno, nota in dictAlunos.items():
  if nota >= 7:
    aprovados.append(aluno)

print(aprovados)
```

```python
#Exercício 4
#1. Crie um programa que receba uma string do usuário.
#2. Imprima a frequência de cada caractere na string.
#Ex: “Unisenai”
#U: 1
#n: 2
#i: 2
#s: 1

string = input('Digite uma string: ')

dict1 = {}

for ch in string:
  if ch in dict1:
    dict1[ch] += 1
  else:
    dict1[ch] = 1

print(dict1)
```

```python
#Exercício 5
#1. Crie um arquivo em que cada linha deverá conter o nome do
#aluno e sua nota, separados por vírgula.
#2. Crie um programa que seja capaz de ler essas informações e
#guardá-las em um dicionário.
#3. O nome do aluno deverá aparecer mais de uma vez no
#arquivo.
#4. Imprima a média do aluno, com base na quantidade de notas
#que apareceram no arquivo. Imprima a quantidade de notas
#junto com a média.

dict1 = {}
dict2 = {}
with open('exercicio05.txt', 'r') as arq:
  for linha in arq:
    linha.replace('\n', '')
    resultado = linha.split(',')
    nome = resultado[0]
    nota = float(resultado[1])

    if nome in dict1:
      dict1[nome] = dict1[nome] + nota
      dict2[nome] = dict2[nome] + 1
    else:
      dict1[nome] = nota
      dict2[nome] = 1

for nome in dict1.keys():
  soma_das_notas = dict1[nome]
  qnt_notas = dict2[nome]
  media = soma_das_notas / qnt_notas
  s = ''
  if qnt_notas > 1:
    s = 's'

  print(f'aluno(a): {nome}, media: {media:.2f} com {qnt_notas} nota{s}')
```
