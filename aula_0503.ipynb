```python
#Entrada
temperatura = "-2"
calibracao:float = 1.0

temperatura_numerica = float(temperatura)
temperatura_calibrada = temperatura_numerica + calibracao

if temperatura_calibrada < 2.0:
  print("Temperatura inferior ao limite!")

elif temperatura_calibrada >= 2.0 and temperatura_calibrada >= 8.0:
  print("Temperatura adequada!")

else:
  print("Temperatura muito alta!")
```

```python
#Exercício 02: Calculadora de Erro de Predição

previsao = 150
resultado = 142

erro_absoluto = previsao - resultado
erro_ao_quadrado = erro_absoluto ** 2
modulo = resultado % 10

print("Erro Absoluto: ", erro_absoluto, " Erro Elevado ao Quadrado: ", erro_ao_quadrado, " Resto da Divisão: ", modulo)
```

```python
#Exercício 04: Classificador de Dados de Sensor

umidade_txt = input("Qual o valor da umidade do solo? ")
umidade = float(umidade_txt)

if umidade < 30:
  print("Ligar irrigação!")

elif umidade >= 30 and umidade <= 60:
  print("Umidade ideal!")

elif umidade > 60:
  print("Alerta: Solo Encharcado!")
```
