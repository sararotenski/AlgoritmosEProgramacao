```python
#Desafio 1
#Implemente uma função chamada buscar, que recebe duas strings origem e subs, devolvendo o índice
#da primeira ocorrência de subs em origem. Se não encontrar, deve retornar -1
#Obs: não use a função pronta “find” do python. O desafio é fazer sua própria versão.
#Ex: buscar(“dificilmasviavel”, “mas”) deve retornar 7
#buscar(“dificilmasviavel”, “masz”) deve retornar -1

def buscar(origem, subs):
    for i in range(len(origem) - len(subs) + 1):
        encontrou = True
        for j in range(len(subs)):
            if origem[i + j] != subs[j]:
                encontrou = False
                break
        if encontrou:
            return i
    return -1

print(buscar("dificilmasviavel", "mas"))
print(buscar("dificilmasviavel", "masz"))
```

```python
#Desafio 2
#Vamos testar a aceleração que um cache em memória pode oferecer. Para isso, imagine que a nossa
#função buscar exige 50 ms para executar, tal como mostrado abaixo:
#Implemente uma função chamada buscar_cache(), capaz de receber a chave e chamar a função
#buscar apenas se necessário.
#Compare o tempo de execução de 1.000 chamadas em “buscar” com “buscar_cache”. Qual o ganho
#% usando o cache?
#Há ganho se as chaves são todas diferentes em buscar_cache?
#Use time.time() para medir o tempo. Essa função retorna o tempo em segundos (+ frações). Basta
#executar antes e depois e pegar a diferença.

import time

def buscar(origem, subs):
    time.sleep(0.05)
    for i in range(len(origem) - len(subs) + 1):
        encontrou = True
        for j in range(len(subs)):
            if origem[i + j] != subs[j]:
                encontrou = False
                break
        if encontrou:
            return i
    return -1

cache = {}

def buscar_cache(origem, subs):
    chave = (origem, subs)
    if chave in cache:
        return cache[chave]
    resultado = buscar(origem, subs)
    cache[chave] = resultado
    return resultado

inicio = time.time()

for i in range(1000):
    buscar("dificilmasviavel", "mas")

fim = time.time()

tempo_sem_cache = fim - inicio

print("Tempo sem cache:", tempo_sem_cache)

inicio = time.time()

for i in range(1000):
    buscar_cache("dificilmasviavel", "mas")
fim = time.time()

tempo_com_cache = fim - inicio
print("Tempo com cache:", tempo_com_cache)

ganho = ((tempo_sem_cache - tempo_com_cache) / tempo_sem_cache) * 100
print(f"Ganho percentual: {ganho:.2f}%")
```

```python
#Desafio 3
#Dada uma string s, encontre o comprimento da maior substring sem caracteres duplicados. Dica: use
#dicionários ou outra estrutura que você conheça (ex hashset).
#Exemplo 1:
#Entrada: s = "abcabcbb"
#Saída: 3
#Explicação: A resposta é "abc", com comprimento 3. Observe que "bca" e "cab" também são
#respostas corretas.
#Exemplo 2:
#Entrada: s = "bbbbb"
#aída: 1
#Explicação: A resposta é "b", com comprimento 1.
#Exemplo 3:
#ntrada: s = "pwwkew"
#Saída: 3
#Explicação: A resposta é "wke", com comprimento 3.
#Observe que a resposta deve ser uma substring; "pwke" é uma subsequência e não uma substring.
#Restrições:
#0 ≤ len(s) ≤ 50.000

def maior_substring(s):
    caracteres = set()
    inicio = 0
    maior = 0
    for fim in range(len(s)):
        while s[fim] in caracteres:
            caracteres.remove(s[inicio])
            inicio += 1
        caracteres.add(s[fim])
        tamanho = fim - inicio + 1
        if tamanho > maior:
            maior = tamanho
    return maior

print(maior_substring("abcabcbb"))
print(maior_substring("bbbbb"))
print(maior_substring("pwwkew"))
```
