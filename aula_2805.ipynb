```python
#exercício 02
#Escreva um código capaz de plotar a função f(x) = x² - 4
#- Crie uma função que recebe x e devolve f(x)
#- Chame a função para vários valores de x e plote a curva
#usando o matplotlib
def func(x):
  return x * x - 4

x = []
y = []
for i in range(-20, 20):
  x.append(i)
  y.append(func(i))

plt.plot(x, y)
plt.show()
```

```python
#adicionando rótulos, títulos, legenda, definindo a cor e estilo,
#salvando a figura, alterando limite dos eixos
def func(x):
  return x * x - 4

x = []
y = []
for i in range(-20, 20):
  x.append(i)
  y.append(func(i))

plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('gráfico de uma função quadrática')

plt.plot(x, y, label = 'curva vermelha', color = 'red', linestyle = 'dashdot')
plt.ylim(-100, 100)
plt.legend()
plt.show()
plt.savefig('teste.png')
```

```python
#plotando duas curvas
x = []
y = []
y2 = []
for i in range(-10, 11):
  x.append(i)
  y.append(func(i))
  y2.append(i)

plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('gráfico de duas funções')

plt.plot(x, y, label = 'curva vermelha', color = 'red', linestyle = 'dashdot')
plt.plot(x, y2, label = 'curva verde', color = 'green', linestyle = 'dotted')
plt.legend()
plt.show()
```

```python
#exercício 04
#Vamos usar o matplotlib para aprender sobre conversão de sinais
#digital – analógico.
#Escreva uma função capaz de gerar a coordenada X, com valores
#espaçados linearmente de 0 a 2pi. Use linspace da biblioteca
#numpy para definir o número de pontos entre 0 e 2pi.
import numpy as np
def angulos(num_angulos):
  x = np.linspace(0, 2 * math.pi, num_angulos)
  return x

x = angulos(8)
y = []
for angulo in x:
  y.append(math.sin(angulo))

plt.plot(x, y)
plt.scatter(x, y)
plt.show()
```

```python
#dois gráficos ao mesmo tempo
import matplotlib.pyplot as plt
import numpy as np
import math

def eixo_x(num_pontos: int):
  return np.linspace(0, 2 * math.pi, num_pontos)

def eixo_y(x: list[float]) -> list[float]:
  resultado = []
  for valor in x:
    resultado.append(math.sin(valor))
  return resultado

def eixo_y_cos(x: list[float]) -> list[float]:
  resultado = []
  for valor in x:
    resultado.append(math.cos(valor))
  return resultado

x = eixo_x(16)
y = eixo_y(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(x, y, marker='*')
axes[0].plot(x, y)
axes[0].set_title('seno')

y2 = eixo_y_cos(x)
axes[1].scatter(x, y2, marker='o', color='orange')
axes[1].plot(x, y2, color='orange')
axes[1].set_title('cosseno')

plt.tight_layout()
plt.show()
```
