```python
#1. Use o método input para ler do teclado e informe se o número lido é positivo ou negativo

numero = int(input('Digite um número: '))

if numero > 0:
  print(numero, ' é positivo!')
elif numero < 0:
  print(numero, ' é negativo!')
else:
  print('Digite um número diferente de zero!')
```

```python
#2. Use o método input para ler do teclado e informe se o número lido é par ou ímpar

numero = int(input('Digite um número: '))

if numero % 2 == 0:
  print(numero, ' é par!')
elif numero % 2 != 0:
  print(numero, ' é ímpar!')
```

```python
#3. Use o método input para ler dois números do teclado e informe qual número é o maior

numeroUm = int(input('Digite um número: '))
numeroDois = int(input('Digite outro número: '))

if numeroUm > numeroDois:
  print(numeroUm, 'é maior do que', numeroDois)
elif numeroDois > numeroUm:
  print(numeroDois, ' é maior do que ', numeroUm)
elif numeroUm == numeroDois:
  print('Esse números são iguais!')
```

```python
#5. Conversor de temperatura: peça ao usuário uma temperatura em Celsius e converta para Fahrenheit
#FÓRMULA: F = (C * 9/5) + 32

celsius = int(input('Digite uma temperatura em Celsius: '))
fahrenheit = (celsius * 9/5) + 32

print('A temperatura', celsius, 'graus em Celsius é', fahrenheit, 'graus em fahrenheit!')
```

```python
#7. Contador até N: peça um número N e mostre todos os números de 1 até N

numero = int(input('Digite um número: '))
i = 1

while i <= numero:
  print(i)
  i = i + 1
```

```python
#8. Refaça o exercício 7, apresentando todos os números separados por espaço

numero = int(input('Digite um número: '))
i = 1

while i <= numero:
  print(i, end=" ")
  i = i + 1
```

```python
#9. Refaça o exercício 8, apresentando todos os números separados por vírgula. Não pode haver vírgula no final

numero = int(input('Digite um número: '))
i = 1

while i <= numero:
    if i < numero:
        print(i, end=", ")
    else:
        print(i, end="")
    i = i + 1
```

```python
#10. Calculadora simples: peça dois números, uma operação e mostre seu resultado

numeroUm = int(input('Digite um número: '))
numeroDois = int(input('Digite outro número: '))
operacao = input('Escolha uma operação básica da matemática: ')

if operacao == 'soma':
  resultado = numeroUm + numeroDois
  print('A soma entre', numeroUm, 'e', numeroDois, 'é', resultado)
elif operacao == 'subtração':
  resultado = numeroUm - numeroDois
  print('A subtração entre', numeroUm, 'e', numeroDois, 'é', resultado)
elif operacao == 'multiplicação':
  resultado = numeroUm * numeroDois
  print('A multiplicação entre', numeroUm, 'e', numeroDois, 'é', resultado)
elif operacao == 'divisão':
  resultado = numeroUm / numeroDois
  print('A divisão entre', numeroUm, 'e', numeroDois, 'é', resultado)
```

```python
#11. Calculadora de média: peça ao usuário que informe valores. O usuário poderá informar quantos valores desejar. Quando o usuário não informar um valor, o programa deverá parar e calcular a média sobre os valores já informados

soma = 0
quantidade = 0

while True:
    valor = input("Digite um valor (ou pressione Enter para parar): ")

    if valor == "":
        break

    soma += float(valor)
    quantidade += 1

if quantidade > 0:
    media = soma / quantidade
    print("Média:", media)
else:
    print("Nenhum valor foi informado.")
```

```python
#12. Filtragem dos ímpares: peça ao usuário que informe valores. O usuário poderá informar quantos valores desejar. Quando o usuário não informar um valor, o programa deverá parar e exibir apenas os número pares

pares = []

while True:
    valor = input("Digite um valor (ou pressione Enter para parar): ")

    if valor == "":
        break

    numero = int(valor)

    if numero % 2 == 0:
        pares.append(numero)

print("Números pares:", pares)
```

```python
#13. Detector de repetição: peça ao usuário que informe valores. O usuário poderá informar quantos valores desejar. Quando o usuário não informar um valor, o programa deverá mostrar o primeiro número repetido

numeros = []

while True:
    valor = input("Digite um valor (ou pressione Enter para parar): ")

    if valor == "":
        break

    numeros.append(int(valor))

repetido = None
vistos = []

for numero in numeros:
    if numero in vistos:
        repetido = numero
        break
    vistos.append(numero)

if repetido is not None:
    print("Primeiro número repetido:", repetido)
else:
    print("Nenhum número repetido foi informado.")
```

```python
#14. Detector de repetição II: peça ao usuário que informe valores. O usuário poderá informar quantos valores desejar. Quando o usuário não informar um valor, o programa deverá mostrar todos os números repetidos

numeros = []

while True:
    valor = input("Digite um valor (ou pressione Enter para parar): ")

    if valor == "":
        break

    numeros.append(int(valor))

vistos = []
repetidos = []

for numero in numeros:
    if numero in vistos:
        if numero not in repetidos:
            repetidos.append(numero)
    else:
        vistos.append(numero)

if repetidos:
    print("Números repetidos:", " ".join(map(str, repetidos)))
else:
    print("Nenhum número repetido foi informado.")
```

```python
#15. Menu interativo: vamos construir um menu simples. Quando o usuário digitar uma opção do menu você deverá informar o texto “Opção X escolhida”, onde X é a opção escolhida. Use um número ou letra (a seu critério) para o usuário sair do menu. Se o usuário digitar algo errado, retorne: opção inválida e apresente novamente o menu

while True:
    print("\n1 - Somar")
    print("2 - Subtrair")
    print("3 - Multiplicar")
    print("4 - Dividir")
    print("s - Sair")

    opcao = input("Opção: ")

    if opcao == "1":
        print("Opção Somar escolhida.")
    elif opcao == "2":
        print("Opção Subtrair escolhida.")
    elif opcao == "3":
        print("Opção Multiplicar escolhida.")
    elif opcao == "4":
        print("Opção Dividir escolhida.")
    elif opcao.lower() == "s":
        print("Saindo...")
        break
    else:
        print("Opção inválida")
```

```python
#16. Calculadora com menu interativo: vamos construir a calculadora do exercício 5 usando um menu simples. Quando o usuário digitar uma opção do menu você deverá solicitar os números e realizar a operação. Se o usuário digitar algo errado, retorne: opção inválida e apresente novamente o menu

while True:
    print("\n1 - Somar")
    print("2 - Subtrair")
    print("3 - Multiplicar")
    print("4 - Dividir")
    print("s - Sair")

    opcao = input("Opção: ")

    if opcao.lower() == "s":
        print("Saindo...")
        break

    if opcao in ["1", "2", "3", "4"]:
        num1 = float(input("Número 1: "))
        num2 = float(input("Número 2: "))

        if opcao == "1":
            resultado = num1 + num2
        elif opcao == "2":
            resultado = num1 - num2
        elif opcao == "3":
            resultado = num1 * num2
        elif opcao == "4":
            if num2 != 0:
                resultado = num1 / num2
            else:
                print("Erro: divisão por zero")
                continue

        print("Resultado:", resultado)

    else:
        print("Opção inválida")
```
