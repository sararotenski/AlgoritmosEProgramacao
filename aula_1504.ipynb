```python
#Exercício 4
#Crie um arquivo qualquer com o nome “arquivo1.txt”. Escreva algo, salve e feche o arquivo.
#Escreva um programa que lê o arquivo acima e informe quantas letras “a” temos no arquivo.
#Utilize with para fechar o arquivo automaticamente.

with open("teste.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()
    quantidade_a = conteudo.count("a")

print(f"Quantidade de letras 'a': {quantidade_a}")
```

```python
#Exercício 5
#Escreva um programa que permita ao usuário inserir dados em
#um arquivo. Use “w” como modo de escrita.
#O programa deverá terminar quando o usuário pressionar enter
#sem informar dados. Nesse caso, abra o arquivo como somente
#leitura e imprima seu conteúdo.

string = input('Digite algo: ')

with open('teste.txt', 'w') as arquivo:

    while string != '':
      arquivo.write(string + '\n')
      string = input('Digite algo: ')

with open('teste.txt', 'r') as arquivo:
  dados = arquivo.read()
  print(dados)
```
