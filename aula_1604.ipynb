```python
#Exercício 6
#Escreva um programa que permita ao usuário inserir dados em
#um arquivo “aos poucos”. Use “a” como modo de escrita.
#O programa deverá abrir o arquivo apenas para salvar o dado,
#fechando-o logo que salvá-lo.
#O programa deverá terminar quando o usuário pressionar enter
#sem informar dados. Nesse caso, abra o arquivo como somente
#leitura e imprima seu conteúdo.

r = True

while True:
  with open('teste.txt', 'a') as arq:
    entrada = input('Digite algo: ')
    if entrada == '':
      break
    else:
      with open('teste.txt', 'a') as arq:
        arq.write(entrada + '\n')
```

```python
#Exercício 8
#Escreva um programa que copie todo o conteúdo de um arquivo
#texto para outro arquivo.
#Use o modo binário para realizar essa cópia, garantindo que o
#arquivo possa ser copiado em qualquer formato.
#Teste o programa acima com um arquivo PDF.

def ler_arquivo(nome: str) -> str:
  with open(nome, 'rb') as arq1:
    return arq1.read()

def escrever_arquivo(nome, dados):
  with open(nome, 'wb') as arq2:
    arq2.write(dados)

dados = ler_arquivo('Aula 08 - Manipulação de Arquivos.pdf')
print(dados)
escrever_arquivo('exercicio8.pdf', dados)
```

```python
#Exercício 10
#Escreva um programa capaz de ler uma lista de números no
#formato CSV e armazená-la em uma lista.
#Imprima o maior número, o menor número e a média dos
#números.

entrada = input('Digite números separados por vírgula: ')

lista_str = entrada.split(',')

numeros = []
for num in lista_str:
    numeros.append(int(num))

maior = max(numeros)
menor = min(numeros)
media = sum(numeros) / len(numeros)

print('Maior número:', maior)
print('Menor número:', menor)
print('Média:', media)
```
