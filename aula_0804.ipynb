```python
#Exercício 01
#Escreva um programa capaz de ler uma string e imprimir o
#primeiro e o último caractere.
#Use len() para descobrir o tamanho da string.

string = input('insira uma string: ')

tamanho = len(string)
print(string[0])
print(string[tamanho - 1])
```

```python
#Exercício 02
#Escreva um programa capaz de ler uma string e imprimir o
#primeiro e o último caractere. Use o índice negativo.

string = input('insira uma string: ')

print(string[0])
print(string[-1])
```

```python
#Exercício 03
#Escreva um programa capaz de ler uma string e imprimi-la na
#vertical (cada caractere deverá estar em uma linha).
#Use while para iterar sobre cada caractere da string.

string = input('insira uma string: ')

indice = 0

while indice < len(string):
  print(string[indice])
  indice = indice + 1
```

```python
#Exercício 04
#Escreva um programa capaz de ler uma string e imprimi-la na
#vertical (cada caractere deverá estar em uma linha).
#Use for para iterar sobre cada caractere da string.

string = input('insira uma string: ')

for caractere in string:
  print(caractere)
```

```python
#Exercício 04

#Escreva um interpretador de Emojis, capaz de converter os seguintes
#emojis:

entrada = input('Digite o emoji: ')

if (entrada[0] == ':'):
  if (entrada[1] == ')'):
    print('😀')
  elif (entrada[1] == '('):
    print('😔')
  else:
    print(entrada)
else:
  print(entrada)
```

```python
#Exercício 05
#Crie uma string e concatene um caractere a ela. Imprima o
#endereço dela antes e após a concatenação. Observe que o
#endereço mudou.
#Obs: use id() para retornar o endereço de memória da string.

strl = 'aaaa'

print(id(strl))

strl = 'aaaa' + 'b'

print(id(strl))
```

```python
#Manipulação de Strings: len()
#Conta o número de caracteres da string

strl = 'aaaa'
print(len(strl))
```

```python
#Manipulação de Strings: .replace()
#Substitui um caractere por outro

str = 'aaaa'
str2 = str.replace('a', 'b')
print(str2)
```

```python
#Manipulação de Strings: .find()
#Indica o índice inicial da palavra

var1 = 'Estou aprendendo a programar em Python'

pos = var1.find('Python')

print(pos)
```

```python
#Manipulação de Strings: .count()
#Conta quantas vezes uma determinada string aparece

strl = 'Estamos aprendendo a linguagem de programação Python, aprendendo?'
qnt = strl.count('aprendendo')
print(qnt)
```

```python
#Manipulação de Strings: upper()
#Coloca todos os caracteres em capslock

strl = 'Estamos aprendendo a linguagem de programação Python, aprendendo?'
print(strl.upper())
```

```python
#Manipulação de Strings: lower()
#Coloca todos os caracteres em letra minúscula

strl = 'Estamos aprendendo a linguagem de programação Python, aprendendo?'
print(strl.lower())
```

```python
#Manipulação de Strings: split()
#Coloca cada conjunto de caractere entre aspas e separa por vírgula

strl = 'Estamos aprendendo a linguagem de programação Python, aprendendo?'
print(strl.split())
```

```python
#Exercício 06
#Crie um programa que recebe uma string como entrada do
#usuário e imprime a string ao contrário.

strl = input('Digite uma frase: ')

for i in range(1, len(strl) + 1):
  print(strl[-i], end='')
```

```python
#Exercício 07
#Crie um programa que recebe uma entrada separada por
#espaços e imprime a frase invertendo as palavras.

strl = input('Digite uma frase: ')

palavras = strl.split(' ')

for i in range(1, len(palavras) + 1):
  print(palavras[-i], end=' ')
```

```python
#Exercício 08: Interpolação de Strings

nome = 'Sara'
idade = 18

print(f'Olá, {nome}! Você tem {idade} anos.')
print('Olá, %s! Você tem %d anos.' % (nome, idade))
print('Olá, {}! Você tem {} anos.'. format(nome, idade))
```

```python
#Fazendo um contador sem count

varl = 'Estou aprendendo a programar em Python'

def count (varl, caractere):
  qnt = 0
  for c in varl:
    if c == caractere:
      qnt += 1
  return qnt

print('O a apareceu:', count(varl, 'a'), 'vezes.')
```
