```python
# Desafio 01: Dado uma lista de números inteiros e um número inteiro X, escreva um programa que retorna os índices dessa lista cuja soma dá X.

def encontrar_indices(lista, x):
  mapa={}

  for i, num in enumerate(lista):
    complemento = x - num

    if complemento in mapa:
      return [mapa[complemento], i]

    mapa[num] = i

lista = [1, 2, 4, 7, 11, -1, 5, -7]
x = 3

print(encontrar_indices(lista, x))
```

```python
# Desafio 02: Implemente uma ordenação bubble-sort

def bubble_sort(lista):
    n = len(lista)

    for pointer1 in range(n):
        for pointer2 in range(0, n - 1):
            if lista[pointer2] > lista[pointer2 + 1]:
              lista[pointer2], lista[pointer2 + 1] = lista[pointer2 + 1], lista[pointer2]

    return lista

lista = [5, 2, 9, 1, 5, 6]
print(bubble_sort(lista))
```
