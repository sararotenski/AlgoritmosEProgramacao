```python
import numpy as np
import pandas as pd
import random

simbolos = ['~', '^', '#', '*', '+']
matriz = []

for i in range(15):
    linha = []
    for j in range(15):
        linha.append(random.choice(simbolos))
    matriz.append(linha)

arquivo_entrada = pd.DataFrame(matriz)

print('Bem-vindo(a) ao Simulador de Sonhos!')
print('\nSeu desafio é chegar na rodada 10 com mais sonhos bons do que pesadelos.')

for rodada in range(1, 11):

    print(f'\n===== RODADA {rodada} =====')
    print(arquivo_entrada)

    print('\nEscolha uma posição para transformar em Sonho Bom (+)')
    print('Digite -1 para passar o turno')

    linha = int(input('Linha: '))

    if linha != -1:
        coluna = int(input('Coluna: '))
        arquivo_entrada.iloc[linha, coluna] = '+'

    copy = arquivo_entrada.copy()

    for i in range(15):
      for j in range(15):
        sonhos_bons = 0
        pesadelos = 0
        lucidos = 0
        dormindo = 0

        for x in range(i - 1, i + 2):
            for y in range(j - 1, j + 2):

                if x == i and y == j:
                    continue

                if 0 <= x < 15 and 0 <= y < 15:

                    vizinho = arquivo_entrada.iloc[x, y]

                    if vizinho == '+':
                        sonhos_bons += 1

                    elif vizinho == '#':
                        pesadelos += 1

                    elif vizinho == '*':
                        lucidos += 1

                    elif vizinho == '~':
                        dormindo += 1

        estado = arquivo_entrada.iloc[i, j]

        # Regra 1 - Se uma célula Dormindo (~) tiver 2 ou mais Sonhos Bons (+) ao redor, ela se transforma em Sonho Bom (+).

        if estado == '~' and sonhos_bons >= 2:
            copy.iloc[i, j] = '+'

        # Regra 2 - Se uma célula Dormindo (~) tiver 2 ou mais Pesadelos (#) ao redor, ela se transforma em Pesadelo (#).

        elif estado == '~' and pesadelos >= 2:
            copy.iloc[i, j] = '#'

        # Regra 3 - Se uma célula Sonho Bom (+) tiver 3 ou mais Sonhos Bons (+) ao redor, ela se transforma em Sonho Lúcido (*).

        elif estado == '+' and sonhos_bons >= 3:
            copy.iloc[i, j] = '*'

        # Regra 4 - Se uma célula Pesadelo (#) tiver 3 ou mais Sonhos Bons (+) ao redor, ela se transforma em Sonho Bom (+).

        elif estado == '#' and sonhos_bons >= 3:
            copy.iloc[i, j] = '+'

        # Regra 5 - Se uma célula Sonho Lúcido (*) tiver menos de 2 vizinhos, ela se transforma em Acordado (^).

        elif estado == '*' and (sonhos_bons + pesadelos + lucidos) < 2:
            copy.iloc[i, j] = '^'

        # Regra 6 - Se uma célula Acordado (^) tiver pelo menos 2 células Dormindo (~) ao redor, ela volta a ser Dormindo (~).

        elif estado == '^' and dormindo >= 2:
            copy.iloc[i, j] = '~'

    arquivo_entrada = copy.copy()

sonhos_bons = 0

for i in range(15):
    for j in range(15):
        if arquivo_entrada.iloc[i, j] == '+':
            sonhos_bons += 1

if sonhos_bons >= 113:
    print('\n✨ O mundo dos sonhos foi salvo!')
    print(f'Quantidade de sonhos bons: {sonhos_bons}')
    print('Você venceu!')
else:
    print('\n🌑 Os pesadelos dominaram o mundo dos sonhos...')
    print(f'Quantidade de sonhos bons: {sonhos_bons}')
    print('Você perdeu!')

arquivo_entrada.to_csv(
    'mundo_final.txt',
    sep=' ',
    header=False,
    index=False
)
```
