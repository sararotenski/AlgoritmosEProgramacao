```python
# 1. Crie uma função capaz de calcular o desconto do INSS.
# 2. Crie uma função capaz de calcular o Imposto de Renda.

def calcular_inss(salario: float) -> float:
  inss = 0

  if salario > 0:
    valor = min(salario, 1412)
    inss += valor * 0.075

  if salario > 1412:
    valor = min(salario, 2666.68) - 1412
    inss += valor * 0.09

  if salario > 2666.68:
    valor = min(salario, 4000.03) - 2666.68
    inss += valor * 0.12

  if salario > 4000.03:
    valor = min(salario, 7786.02) - 4000.03
    inss += valor * 0.14

  return inss

def calcular_irpf(salario: float) -> float:
    if salario <= 2428.80:
        return 0

    elif salario <= 2826.65:
        return salario * 0.075 - 182.16

    elif salario <= 3751.05:
        return salario * 0.15 - 394.16

    elif salario <= 4664.68:
        return salario * 0.225 - 675.49

    else:
        return salario * 0.275 - 908.73

salario = 4500
desconto_inss = calcular_inss(salario)
desconto_irpf = calcular_irpf(salario - desconto_inss)
salario_liquido = salario - desconto_inss - desconto_irpf

print('INSS:', round(desconto_inss, 2))
print('IRPF:', round(desconto_irpf, 2))
print('Salário Líquido:', round(salario_liquido, 2))
```

```python
nomes = [
    "João", "Pedro", "Lucas", "Mateus", "Gabriel",
    "Maria", "Ana", "Juliana", "Fernanda", "Patrícia",
    "Rafael", "Bruno", "Carlos", "Felipe", "Marcos",
    "Mariana", "Amanda", "Letícia", "Renata", "Carla",
    "André", "Rodrigo", "Diego", "Fernando", "Ricardo",
    "Camila", "Aline", "Larissa", "Beatriz", "Carolina"
]

salarios = [
    15432, 1981, 1654, 11234, 1876,
    8450, 13210, 4654, 3321, 1999,
    2002, 11765, 19201, 3450, 1789,
    12100, 10456, 990, 970, 24032,
    5888, 2345, 7777, 5001, 20000,
    18800, 12666, 14700, 1101, 5000
]

# Imagine que os dados acima foram extraídos de uma planilha do RH

# Há quantos funcionários nessa empresa?
numero_funcionarios = len(nomes)
print('Há', numero_funcionarios, 'funcionários na empresa')

# Qual é o salário médio?
media_salarial = sum(salarios) / len(salarios)
print('A média salarial é', media_salarial)

# Quantas pessoas ganham no mínimo 20% acima da média?
salario_mais20 = media_salarial * 1.2

contador = 0

for salario in salarios:
  if salario >= salario_mais20:
    contador += 1

print("Limite (20% acima):", round(salario_mais20, 2))
print("Quantidade de pessoas:", contador)

# Quantas pessoas ganham no mínimo 20% abaixo da média?
salario_menos20 = media_salarial * 0.8

contador = 0

for salario in salarios:
  if salario <= salario_menos20:
    contador += 1

print('Limite (20% a menos):', round(salario_menos20, 2))
print('Quantidade de pessoas:', contador)

# Quem é a pessoa que tem o maior salário?
indice_maior = salarios.index(max(salarios))
pessoa_maior = nomes[indice_maior]
salario_maior = salarios[indice_maior]

print("Pessoa com maior salário:", pessoa_maior)
print("Maior salário:", salario_maior)

# Quem é a pessoa que tem o menor salário?
indice_menor = salarios.index(min(salarios))
pessoa_menor = nomes[indice_menor]
salario_menor = salarios[indice_menor]

print("Pessoa com menor salário:", pessoa_menor)
print("Menor salário:", salario_menor)

# Qual é o tamanho da folha de pagamentos? (considerem 13 meses de salário e 1/3 de férias, desconsiderem impostos)
folha = 0

for salario in salarios:
    folha += salario * 13 + salario / 3

print("Tamanho da folha de pagamentos: R$ {:.2f}".format(folha))

# Qual é o salário anual de cada funcionário, incluíndo 13o e 1/3 de férias?
salario_anual = []

for salario in salarios:
    total = salario * 13 + salario / 3
    salario_anual.append(total)

for nome, total in zip(nomes, salario_anual):
    print(f"{nome}: R$ {total:.2f}")
```
