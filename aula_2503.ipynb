```python
def meu_metodo(parametro1: int, parametro2: float):
  if not isinstance(parametro1, int):
    print('Parametro 1 incorreto')
  retorno = parametro1 + parametro2
  return retorno

resultado = meu_metodo('a', 'b')

print(resultado)
```

```python
def exponencial (a: float, b: float = 2) -> float:
  return a ** b

print(exponencial(3, 2))
```

```python
def soma(a, b):
  return a + b

def subtracao(a, b):
  return a - b

def multiplicacao(a, b):
  return a * b

def divisao(a, b):
  return a / b

def quatro_operacoes(a, b):
  op1 = soma(a, b)
  op2 = subtracao(a, b)
  op3 = multiplicacao(a, b)
  op4 = divisao(a, b)
  return op1, op2, op3, op4

resultado = quatro_operacoes(3, 3)
print(resultado)
```

```python
def circulo(raio):
  comprimento = 2 * 3.1415 * raio
  area = 3.1415 * raio * raio
  return comprimento, area

comprimento, area = circulo(20)

print('Comprimento:', comprimento)
print('Área:', area)
```

```python
def circulo(raio, comprimento: bool = True, area: bool = False):
  comp = 2 * 3.1415 * raio
  areaValor = 3.1415 * raio * raio

  if comprimento == True and area == True:
    return comp, areaValor
  elif comprimento == True and area == False:
    return comp, None
  elif comprimento == False and area == True:
    return None, areaValor
  else:
    return None, None

comp, areaValor = circulo(20, comprimento = False, area = True)

print('Comprimento:', comp)
print('Área:', areaValor)
```

```python
def fibonacci_recursive(n):
  if n <= 1:
    return n
  else:
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

def fib(num_elementos):
  print('Fibonacci:')
  for i in range(num_elementos):
    print(fibonacci_recursive(i))

fib(2)
```
