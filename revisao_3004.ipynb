```python
#1. Escreva um programa que leia a idade de uma pessoa e informe:
#• “Criança” (0–12)
#• “Adolescente” (13–17)
#• “Adulto” (18–59)
#• “Idoso” (60+)
idade = int(input('Digite sua idade: '))

if idade <= 12:
  print('Você é criança!')
elif idade > 12 and idade < 18:
  print('Você é adolescente!')
elif idade >= 18 and idade < 60:
  print('Você é adulto!')
elif idade >= 60:
  print('Você é idoso!')
```

```python
#3. Faça um programa que receba três números e mostre o maior deles.
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))
n3 = int(input('Digite mais um número: '))

if n1 > n2 and n1 > n3:
  print(f'{n1} é o maior número!')
elif n2 > n1 and n2 > n3:
  print(f'{n2} é o maior número!')
elif n3 > n2 and n3 > n1:
  print(f'{n3} é o maior número!')
```

```python
#Mude o exercício anterior para função.
def maior(a, b, c):
  if a > b and a > c:
    return(a)
  elif b > a and b > c:
    return(b)
  elif c > b and c > a:
    return(c)

print(maior(55, 9, 8))
```

```python
#4. Escreva um programa que informe se um número é par e positivo ao mesmo tempo.
num = int(input('Digite um número: '))

if num % 2 == 0 and num > 0:
  print(f'{num} é um número par e positivo!')
elif num % 2 == 0 and num < 0:
  print(f'{num} é um número par e negativo!')
elif num % 2 != 0 and num > 0:
  print(f'{num} é um número ímpar e positivo!')
elif num % 2 != 0 and num < 0:
  print(f'{num} é um número ímpar e negativo!')
```

```python
#7. Faça um programa que calcule a soma de todos os números de 1 até N (entrada do usuário).
n = int(input('Digite um número: '))
soma = 0

for i in range(1, n + 1):
  soma += i
print(f'A soma dos números de 1 até {n} é {soma}.')
```

```python
#8. Escreva um programa que peça números até o usuário digitar 0 e então mostre a soma deles.
num = int(input('Digite um número: '))
soma = 0

while num != 0:
  soma += num
  num = int(input('Digite um número: '))
print(f'A soma dos números digitados é {soma}.')
```

```python
#10. Crie uma função que receba base e altura e retorne a área de um triângulo.
base = int(input('Digite a base do triângulo: '))
altura = int(input('Digite a altura do triângulo: '))

def area(base, altura):
  return (base * altura) / 2

print(f'A área do triângulo é {area(base, altura)}')
```

```python
#11. Faça uma função que receba uma lista de números e retorne o maior valor.
def maior(lista):
  return max(lista)

nums = [2, 5, 7, 3, 8, 19, 6, 55, 160]
print(maior(nums))
```

```python
#12. Cria um código que recebe 10 números do usuário e guarda em uma lista apenas os maiores que 50.
lista = []

for i in range(10):
  num = int(input('Digite um número: '))
  if num > 50:
    lista.append(num)

print(lista)
```

```python
#13. Conte quantas vogais existem em uma frase digitada.
frase = input('Digite uma frase: ').lower()
vogais = 'aeiouáéíóúâêîôûãõ'
i = 0

for letra in frase:
  if letra in vogais:
    i += 1

print(f'A frase contém {i} vogais.')
```

```python
#15. Crie um programa que recebe N números do usuário. Imprima apenas os números que só foram
#informados uma única vez. O programa deve parar quando o usuário não informar um número e aí
#exibir aqueles que só foram informados uma única vez.
lista = []
while True:
  entrada = input('Digite um número: ')
  if entrada == '':
    break
  numero = int(entrada)
  lista.append(numero)

unicos = [n for n in lista if lista.count(n) == 1]
print(f'Números informados apenas uma vez:')
for n in unicos:
  print(n)
```
