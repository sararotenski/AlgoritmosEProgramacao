```python
#crie uma função que calcula o IMC (peso/altura²)
def imc(peso, altura):
  valor = peso / altura ** 2
  print(valor)

imc(55, 1.55)
```

```python
#crie uma função que aproxima a raiz quadrada usando Newton-Raphson
def aproxima_raiz(p, x):
  resultado = (p + (x * x)) / 2 * x
  return resultado

raiz_aproximada = aproxima_raiz(4.5, 2)
print(raiz_aproximada)
```
