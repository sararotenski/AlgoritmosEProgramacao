```python
#mantém o 3 e transforma os outros em zero
m1 = np.array([
    [1, 2, 3],
    [3, 2, 1],
    [-1, -2, -3]
])
m2 = np.where(m1 == 3, m1, 0)
print(m2)
```

```python
#acrescentando linhas
import numpy as np
m1 = np.full((3, 3,), 10.2)
m2 = np.full((1, 3), 1.0)
m3 = np.vstack((m1, m2))
print(m3)
```

```python
#acrescentando colunas
import numpy as np
m1 = np.full((3, 3,), 10.2)
m2 = np.full((3, 1), 1.0)
m3 = np.hstack((m1, m2))
print(m3)
```

```python
#manipulando imagens
from matplotlib import pyplot as plt
import numpy as np
import cv2
img = cv2.imread('casa.png')
img = np.array(img)
img = cv2. cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

m2 = np.full(img.shape, [0, 250, 0])
print(img.shape)
plt.imshow(img - m2)
plt.axis('off')
plt.show()
```

```python
#juntando duas imagens uma em cima da outra
img = cv2.imread('casa.png')
img = np.array(img)
img2 = img.copy()
img3 = np.vstack((img, img2))
plt.imshow(img3)
plt.axis('off')
plt.show()
```

```python
#juntando duas imagens uma ao lado da outra
img = cv2.imread('casa.png')
img = np.array(img)
img2 = img.copy()
img3 = np.hstack((img, img2))
plt.imshow(img3)
plt.axis('off')
plt.show()
```

```python
#colocando imagens na diagonal
img = cv2.imread('casa.png')
img = np.array(img)
img2 = img.copy()
m1 = np.full(img.shape, [250, 250, 250])
m2 = m1.copy()
lin1 = np.hstack((m1, img))
lin2 = np.hstack((img2, m2))
img3 = np.vstack((lin1, lin2))
plt.imshow(img3)
plt.axis('off')
plt.show()
```
