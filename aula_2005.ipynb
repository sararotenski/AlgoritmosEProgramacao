```python
#Exercício I
#1. Crie uma lista contendo 10 nomes de pessoas
#2. Crie uma lista contendo 10 salários aleatórios, variando entre 0 e
#10.000. Use random para gerar os números
#3. Crie um dicionário juntando as duas informações acima.
#4. Use o dicionário para criar um dataframe
import random

nomes = ['caio', 'sara', 'helo', 'gabi', 'lari', 'danny', 'arthur', 'ana', 'olga', 'richard']
salarios = []
for i in range(10):
  salarios.append(random.random() * 10000)

dados = {
    'nomes': nomes,
    'salarios': salarios
}

df = pd.DataFrame(dados)
df
```

```python
#Exercício 2
#Imprima todos os salários, acessando a coluna correta no dataframe e
#iterando sobre o resultado de .items()
for p in df['salarios'].items():
  print(p)
```

```python
#Exercício 3
#Encontre o maior salário. Use o looping for para verificar qual é o
#maior salário e imprima seu valor e qual é a posição dele na coluna.
maior = 0
idx_maior = -1
for i, valor in df['salarios'].items():
  if maior < valor:
    maior = valor
    idx_maior = i

print(f'O maior salário é {maior} e está na posição {idx_maior}')
```

```python
#acessando um elemento da coluna por índice
col = df['salarios']
print(col[0])
```

```python
#Exercício 4
#A partir do que encontramos no exercício 3…
#Imprima o nome da pessoa com o maior salário. Use a posição
#encontrada no exercício 3 para achar o nome da pessoa.
maior = 0
idx_maior = -1
for i, valor in df['salarios'].items():
  if maior < valor:
    maior = valor
    idx_maior = i

nome = df['nomes'][idx_maior]
print(f'{nome} possui o maior salário: {maior}')
```

```python
#funções úteis presentes na coluna
col = df['salarios']
print(col.max()) #valor máximo
print(col.min()) #valor mínimo
print(col.mean()) #valor médio
print(col.argmax()) #posição do valor máximo
print(col.argmin()) #posição do valor mínimo
print(col.count()) #número de itens
```

```python
#Exercício 5
#Imprima o nome da pessoa com o maior salário. Use os métodos de
#coluna para achar a posição com o maior salário.
indice = col.argmax()
nome = df['nomes'][indice]
print(nome)
```

```python
#Exercício 6
#Imprima os nomes das pessoas cujo salário é abaixo da média.
media = col.mean()
for i in df['salarios']:
  if i < media:
    print(i)
```

```python
#versão com filtro
media = df['salarios'].mean()
filtro = df[df['salarios'] < media]
print(filtro)
```

```python
#Exercício 7
#Vamos criar um arquivo CSV e lê-lo usando pandas.
#1. Crie uma lista de nomes. Ex: nomes = [ “Julia”, “Pedro”, “João”…]
#2. Crie uma lista de nomes de disciplinas. Ex: disciplinas =
#[“matemática”, “física”, “português” …]
#3. Abra um arquivo dados.csv para escrita (w)
#4. Escreva o cabeçalho: nome,disciplina,nota
#5. Gere 1 linha para cada nome x disciplina, criando uma nota
#aleatória com random, de 0 a 10.
nomes = ['sara', 'caio', 'helo']
disciplinas = ['algoritmos', 'banco de dados', 'pre-calculo']

with open('dados.csv', 'w') as arquivo:
  arquivo.write('nome,disciplina,nota\n')
  for nome in nomes:
    for disc in disciplinas:
      nota = random.random() * 10
      print(f'{nome}, {disc}, {nota:.2f}')
      arquivo.write(f'{nome}, {disc}, {nota:.2f}\n')
```

```python
#Exercício 08
##Carregue o arquivo dados.csv em um dataframe e exiba:
#1. O aluno com a maior nota
#2. O aluno com a menor nota
#3. A quantidade de alunos acima da media
#4. A quantidade de alunos abaixo da media
#Obs: use os métodos de coluna para calcular as estatísticas requisitas
#acima.
df = pd.read_csv('dados.csv')

maior_nota = df['nota'].max()
maior_nota_idx = df['nota'].argmax()

menor_nota = df['nota'].min()
menor_nota_idx = df['nota'].argmin()

aluno_maior = df['nome'][maior_nota_idx]
print(f'O aluno {nome} possui a maior nota: {maior_nota}')

aluno_menor = df['nome'][menor_nota_idx]
print(f'O aluno {nome} possui a menor nota: {menor_nota}')

media = df['nota'].mean()
print(df[df['nota'] > media])
print(df[df['nota'] < media])

lista_maior = []
for i in df['nota']:
  if i > media:
    lista_maior.append(i)
print(f'O número de notas acima da média é {len(lista_maior)}')

lista_menor = []
for i in df['nota']:
  if i < media:
    lista_menor.append(i)
print(f'O número de notas abaixo da média é {len(lista_menor)}')
```

```python
#Exercício 09
#Vamos “separar” nossos dados em tabelas, cada uma em um arquivo
#CSV separado e ler tudo usando pandas.
#1. Crie um dicionário de ID e nome de disciplinas. Ex: disciplinas = {1:
#“matemática”, 2: “física”, 3: “português” …}
#2. Crie uma lista de nomes. Ex: nomes = [ “Julia”, “Pedro”, “João”…]
#3. Abra um arquivo chamado materias.csv para escrita e inclua o
#abeçalho codigo_disciplina, nome_disciplina
#4. Para cada item no dicionário do passo 1, inclua uma linha no
#arquivo com “chave,valor” do dicionário (que é o código e o nome
#da disciplina)
disciplinas = {
    1: 'MLOps',
    2: 'business intelligence',
    3: 'banco de dados',
    4: 'big data',
    5: 'computação em nuvem',
    6: 'machine learning'
    }
nomes = ['sara', 'caio', 'helo']

with open('materias.csv', 'w') as arquivo:
  arquivo.write('codigo_disciplina,nome_disciplina\n')
  for chave, valor in disciplinas.items():
    arquivo.write(f'{chave},{valor}\n')
```

```python
#Exercício 10
#Vamos “separar” nossos dados de notas:
#1. Abra um arquivo chamado notas.csv para escrita e inclua o
#cabeçalho aluno,codigo_disciplina,nota.
#2. Para cada nome da lista de nomes e para cada chave do dicionário
#de disciplinas, inclua uma linha no arquivo com o valor
#“nome,chave do dicionário,nota”. Gere a nota usando random de
#0 a 10.
with open('notas.csv', 'w') as arq:
  arq.write('aluno,codigo_disciplina,nota\n')
  for aluno in nomes:
    for codigo_disciplina in disciplinas.keys():
      nota = random.random() * 10
      arq.write(f'{aluno},{codigo_disciplina},{nota:.2f}\n')
```

```python
#Iterando sobre cada linha de um Dataframe
df = pd.read_csv('notas.csv')
for indice, linha in df.iterrows():
  print(linha)
```

```python
#Exercício 12
#Vamos “juntar” nossos dados manualmente:
#1. Abra o arquivo chamado materias.csv como um dataframe
#2. Abra o arquivo chamado notas.csv como um dataframe
#3. Imprima a lista de alunos e suas notas com o formato:
#Aluno, nome da matéria, nota
df_materias = pd.read_csv('materias.csv')
df_materias = df_materias.set_index('codigo_disciplina')
df_notas = pd.read_csv('notas.csv')

for indice, linha in df_notas.iterrows():
  nome_aluno = linha['aluno']
  codigo = linha['codigo_disciplina']
  nota = linha['nota']
  nome_materia = df_materias.loc[codigo, 'nome_disciplina']
  print(f'{nome_aluno}, {nome_materia}, {nota}')
```

```python
#Exercício 13
#Refaça o exercício 12, mas agora imprima a lista de alunos e suas
#notas com o formato:
#Aluno:
#- matemática: 5.6
#- portugues: 7.2
#Obs: para juntarmos as notas do aluno, precisamos usar um
#dicionário!
dic = {
    'aluno': [],
    'disciplina': [],
    'nota': []
    }

for indice, linha in df_notas.iterrows():
  nome = linha['aluno']
  codigo = linha['codigo_disciplina']
  nome_materia = df_materias.loc[codigo, 'nome_disciplina']
  nota = linha['nota']
  dic['aluno'].append(nome)
  dic['disciplina'].append(nome_materia)
  dic['nota'].append(nota)
df = pd.DataFrame(dic)
print(df)
```
