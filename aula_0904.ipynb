```python
#Exercício 02
#Crie um arquivo qualquer com o nome “arquivo1.txt”. Escreva algo, salve e
#feche o arquivo.
#Escreva um programa que lê o arquivo acima e imprima a primeira linha

cursor = open('arquivo1.txt')

print(cursor.readline())

cursor.close()
```

```python
#Modifique o programa para que ele imprima todo o conteúdo do arquivo,
#numerando as linhas.

cursor = open('arquivo1.txt', 'r')

i = 1
for linha in cursor:
  print(f'linha {i}: {linha}')
  i += 1

cursor.close()
```

```python
#Refaça o exercício acima utilizando while e readline

cursor = open('arquivo1.txt', 'r')

i = 1
linha = cursor.readline()

while linha != '':
    print(f'linha {i}: {linha}')
    i += 1
    linha = cursor.readline()

cursor.close()
```

```python
#Exercício 3
#Crie um arquivo qualquer com o nome “arquivo1.txt”. Escreva
#algo, salve e feche o arquivo.
#Escreva um programa que lê o arquivo acima e informe:
#1. o número de linhas do arquivo
#2. o número de caracteres (que não são espaço em branco).
#3. o número de “espaços” em branco.

with open('arquivo1.txt', 'r') as f:
    linhas = 0
    caracteres = 0
    espacos = 0

    for linha in f:
        linhas += 1

        for c in linha:
            if c.isspace():
                espacos += 1
            else:
                caracteres += 1

print("Número de linhas:", linhas)
print("Número de caracteres (sem espaços):", caracteres)
print("Número de espaços em branco:", espacos)

cursor.close()
```
